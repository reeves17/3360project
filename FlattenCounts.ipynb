Import relevant packages

```python
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import json
import ast
```

Create dataframes<br/><br/>
(This code is not fast)

```python
tedTalks = pd.read_csv("ted_main.csv")
transcripts = pd.read_csv("transcripts.csv")

columns = ['Funny','Beautiful','Ingenious','Courageous','Longwinded','Confusing','Informative','Fascinating','Unconvincing','Persuasive','Jaw-dropping','OK','Obnoxious','Inspiring']
flatten = pd.DataFrame(columns=columns)

for index, row in tedTalks.iterrows():
    t = ast.literal_eval(row['ratings'])
    df = pd.DataFrame(json_normalize(t))
    flatten = flatten.append(df.set_index('name').T.iloc[0])

flatten = flatten.reset_index()
flatten = flatten.drop(['index'], axis=1)  
flatten.head(3)
```

```python
flatten.Funny = flatten.Funny.astype('int64')
flatten.Beautiful = flatten.Beautiful.astype('int64')
flatten.Ingenious = flatten.Ingenious.astype('int64')
flatten.Courageous = flatten.Courageous.astype('int64')
flatten.Longwinded = flatten.Longwinded.astype('int64')
flatten.Confusing = flatten.Confusing.astype('int64')
flatten.Informative = flatten.Informative.astype('int64')
flatten.Fascinating = flatten.Fascinating.astype('int64')
flatten.Unconvincing = flatten.Unconvincing.astype('int64')
flatten.Persuasive = flatten.Persuasive.astype('int64')
flatten['Jaw-dropping'] = flatten['Jaw-dropping'].astype('int64')
flatten.OK = flatten.OK.astype('int64')
flatten.Obnoxious = flatten.Obnoxious.astype('int64')
flatten.Inspiring = flatten.Inspiring.astype('int64')
tedTalks = tedTalks.join(flatten)
tedTalks.info()
```
